```python
from glob import glob
import os

# import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pynwb import NWBHDF5IO
```

## Specify location of NWB directory

```python
NWB_file_dir = r"T:\zap-n-zip\eLife_manuscript_NWBfiles\000458"
```

Each folder belongs to a single subject (mouse). Inside the folders there are individual NWB files for each experiment (session) and subjects have 1 or 2 sessions.

## Create a dataframe with all subjects/sessions with some metadata

```python
subject_session_list = []
for filei in glob(os.path.join(NWB_file_dir, '**\*.nwb')):
    io = NWBHDF5IO(filei, mode="r")
    nwbfile = io.read()
    subject_session_list.append([
        nwbfile.subject.subject_id, nwbfile.session_id, nwbfile.subject.sex, filei,
        np.any([True if 'EEG' in x else False for x in nwbfile.electrode_groups.keys()]),
        np.any([True if 'probe' in x else False for x in nwbfile.electrode_groups.keys()]),
    ])
    io.close
subject_df = pd.DataFrame(subject_session_list, columns=['subject', 'session_id', 'sex', 'file_loc', 'has_EEG', 'has_NPX'])
```

### Load EEG

```python
nwbfile.acquisition['ElectricalSeriesEEG']
```

```python
EEGdata = nwbfile.acquisition['ElectricalSeriesEEG'].data[:] * nwbfile.acquisition['ElectricalSeriesEEG'].conversion * 1E6
EEGtimestamps = nwbfile.acquisition['ElectricalSeriesEEG'].timestamps[:]
```

Loading EEG data * conversion * 1E6 loads the EEG data as microvolts. It will be a 2D array (time, electrodes).

```python
print(EEGdata.shape)
```

### Load LFP

LFP is stored in this NWB file here: nwbfile.acquisition["LFPprobeX"]["ElectricalSeriesprobeX"]

```python
choose_probe = 'probeF'
```

```python
nwbfile.acquisition['LFP'+choose_probe]['ElectricalSeries'+choose_probe]
```

```python
nwbfile.acquisition['LFP'+choose_probe]['ElectricalSeries'+choose_probe].data.shape
```

The LFP data is large (384 chs for one probe, sampled at 2.5 kHz). Same as the EEG data, LFP.data[:] * conversion * 1E6 loads the data into memory as microvolts. It will be a 2D array (time, electrodes).

```python
LFPtimestamps = nwbfile.acquisition['LFP'+choose_probe]['ElectricalSeries'+choose_probe].timestamps[:]
print(LFPtimestamps.shape)
```

```python
inds = LFPtimestamps < 120 # choose to load the first 120 s
LFPdata_chunk = nwbfile.acquisition['LFP'+choose_probe]['ElectricalSeries'+choose_probe].data[inds, :] * nwbfile.acquisition['LFP'+choose_probe]['ElectricalSeries'+choose_probe].conversion * 1E6
print(LFPdata_chunk.shape)
```

### Looking at the units

```python
all_units_df.head()
```

The units included in these NWB files are a subselection from the total Kilosort output. We used some of the metrics described here (https://allensdk.readthedocs.io/en/latest/_static/examples/nb/ecephys_quality_metrics.html) as they did in this paper (https://www.nature.com/articles/s41586-020-03171-x#Sec5, see Methods, Spike sorting section). Specifically, we used the following quality metrics and thresholds:
- isi_violations < 0.5
- amplitude_cutoff < 0.1
